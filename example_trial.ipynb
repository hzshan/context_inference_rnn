```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import task_generation as taskgen
```

```python
pro_trial, boundaries = taskgen.make_go_trial(0, 0.01)
labels = ['stim1 cos', 'stim1 sin', 'stim2 cos', 'stim2 sin', 'fixation']

_, axes = plt.subplots(nrows=5, ncols=1, sharex=True)
axes = axes.ravel()
for i in range(5):
    axes[i].plot(pro_trial['s'][i])
    axes[i].set_ylim(-1.2, 1.2)
    for b in boundaries:
        axes[i].axvline(b, color='k', linestyle='--')
    axes[i].set_ylabel(labels[i])
plt.xlabel('timesteps')
```
