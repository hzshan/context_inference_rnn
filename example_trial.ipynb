```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import task_generation, plots

plots.set_rcparams()
```

```python
# choose a subset of tasks to learn
np.random.seed(10)
trial, boundaries = task_generation.compose_trial(
            task_generation.task_dict['PRO_M'],
            {'theta_task': 3}, 0.05, 0.05, p_stay=0.97, min_Te=15, d_stim=np.pi/4)

_, axes = plt.subplots(figsize=(2, 1), sharex=True, nrows=2, ncols=1)
plt.sca(axes[0])
plt.plot(np.arange(len(trial['s'])) + 1, trial['s'], color='C1')
plt.ylim(-1.2, 1.2)
for b in boundaries:
    plt.axvline(b, color='k', linestyle=':')
plt.xlabel('time')
plt.xticks([])

plt.sca(axes[1])
plt.plot(np.arange(len(trial['s'])) + 1, trial['y'], color='C0')
plt.ylim(-1.2, 1.2)
for b in boundaries:
    plt.axvline(b, color='k', linestyle=':')
plt.xticks([])
plt.xlabel('time')

plt.savefig('Figures/example_trial.svg')
```
