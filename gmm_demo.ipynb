```python
import sklearn.mixture
import numpy as np
import matplotlib.pyplot as plt

true_means = np.array([[0, 2], [1.5, -1], [-1.5, -1]]).astype(np.float32)
data = np.repeat(true_means, 30, axis=0)
data += np.random.normal(np.zeros_like(data), scale=0.05)

init_means = np.random.normal(size=(3, 2))
# init_means[1:, :] += 10

GMM = sklearn.mixture.GaussianMixture(n_components=3,
                                      means_init=init_means,
                                      covariance_type='spherical',
                                      verbose=1)
GMM.fit(data)

fig, axes = plt.subplots(figsize=(5, 3), nrows=1, ncols=2)
plt.sca(axes[0])
plt.scatter(data[:, 0], data[:, 1], c='k', alpha=0.5)
plt.xlabel('dim1')
plt.ylabel('dim2')
plt.title('data')
plt.gca().set_aspect(1)

plt.sca(axes[1])
plt.scatter(true_means[:, 0], true_means[:, 1], c='r', label='true means')
plt.scatter(init_means[:, 0], init_means[:, 1], c='g', label='initial means')
plt.scatter(GMM.means_[:, 0], GMM.means_[:, 1], c='b', label='fitted means', marker='x')
plt.gca().set_aspect(1)
plt.xlabel('dim1')
plt.ylabel('dim2')
plt.legend(bbox_to_anchor=(1, 0.5))

plt.tight_layout()
```
